```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import AdaBoostClassifier
```

```python
data=pd.read_csv('30_features.csv',index_col=0)
```

```python
data.head()
```

```python
X=data.iloc[46749:833507,0:30].reset_index(drop=True)
```

```python
#taking a look at the data
X
```

```python
#specifying y 
y=data.iloc[46749:833507,30].reset_index(drop=True)
```

```python
X_oot=data.iloc[833507:,0:30]
```

```python
y_oot=data.iloc[833507:,30]
```

```python
skf = StratifiedKFold(n_splits=10)
```

```python
for train_index, test_index in skf.split(X, y):
    print("TRAIN:", train_index, "TEST:", test_index)
```

```python
clf = AdaBoostClassifier(n_estimators=100, random_state=0)
```

```python
fdr=pd.DataFrame()
```

```python
for train_index, test_index in skf.split(X, y):
    i=1
    X_train, X_test = X.iloc[train_index,:], X.iloc[test_index,:]
    y_train, y_test = y[train_index], y[test_index]
    clf.fit(X_train,y_train)
    X_trn_eval = X_train.copy()
    X_oot_eval = X_oot.copy()
    X_tst_eval = X_test.copy()
    X_trn_eval['predicted']=clf.predict_proba(X_train)[:,1]
    X_trn_eval['Fraud'] = y_train
    topRows = int(round(X_trn_eval.shape[0]*0.03))
    temp = X_trn_eval.sort_values('predicted',ascending=False).head(topRows)
    needed = temp.loc[:,'Fraud']
    fdr.loc[i,'training']=sum(needed)/sum(X_trn_eval.loc[:,'Fraud'])
    
    X_tst_eval['predicted']=clf.predict_proba(X_test)[:,1]
    X_tst_eval['Fraud'] = y_test
    topRows = int(round(X_tst_eval.shape[0]*0.03))
    temp = X_tst_eval.sort_values('predicted',ascending=False).head(topRows)
    needed = temp.loc[:,'Fraud']
    fdr.loc[i,'testing']=sum(needed)/sum(X_tst_eval.loc[:,'Fraud'])
    
    X_oot_eval['predicted']=clf.predict_proba(X_oot)[:,1]
    X_oot_eval['Fraud'] = y_oot
    topRows = int(round(X_oot_eval.shape[0]*0.03))
    temp = X_oot_eval.sort_values('predicted',ascending=False).head(topRows)
    needed = temp.loc[:,'Fraud']
    fdr.loc[i,'oot']=sum(needed)/sum(X_oot_eval.loc[:,'Fraud'])
    i=i+1
    avg_tr=fdr['training'].mean()
    avg_te=fdr['testing'].mean()
    avg_oot=fdrr['oot'].mean()
```

```python
clf1 = AdaBoostClassifier(n_estimators=100, random_state=0,learning_rate=0.01)
clf1.fit(X_train,y_train)
X_trn_eval['predicted']=clf1.predict_proba(X_train)[:,1]
X_trn_eval['Fraud'] = y_train
topRows = int(round(X_trn_eval.shape[0]*0.03))
temp = X_trn_eval.sort_values('predicted',ascending=False).head(topRows)
needed = temp.loc[:,'Fraud']
fdr.loc[i,'training']=sum(needed)/sum(X_trn_eval.loc[:,'Fraud'])
    
X_tst_eval['predicted']=clf1.predict_proba(X_test)[:,1]

X_tst_eval['Fraud'] = y_test

topRows = int(round(X_tst_eval.shape[0]*0.03))

temp = X_tst_eval.sort_values('predicted',ascending=False).head(topRows)

needed = temp.loc[:,'Fraud']

fdr.loc[i,'testing']=sum(needed)/sum(X_tst_eval.loc[:,'Fraud'])
    

X_oot_eval['predicted']=clf1.predict_proba(X_oot)[:,1]

X_oot_eval['Fraud'] = y_oot

topRows = int(round(X_oot_eval.shape[0]*0.03))

temp = X_oot_eval.sort_values('predicted',ascending=False).head(topRows)

needed = temp.loc[:,'Fraud']

fdr.loc[i,'oot']=sum(needed)/sum(X_oot_eval.loc[:,'Fraud'])
avg_tr=fdr['training'].mean()
avg_te=fdr['testing'].mean()
avg_oot=fdrr['oot'].mean()
```
